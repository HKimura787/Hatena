## Minicondaのインストール

[Minicondaのダウンロードページ](https://docs.conda.io/en/latest/miniconda.html)に行き、Linux用のインストーラの名前とURLを確認しておく。

以下のセルに以下のように記載する（**空白厳禁**）

---


- `MINOCONDA_INSTALLER`=`shファイル名`　（とりあえず、最新版）
- `MINICONDA_DOWNLOAD_HP`=`URL`

（最終確認：2021/7/3）

```python
%%bash
MINICONDA_INSTALLER=Miniconda3-py39_4.9.2-Linux-x86_64.sh
MINICONDA_DOWNLOAD_HP=https://repo.anaconda.com/miniconda

MINICONDA_PREFIX=/usr/local
wget $MINICONDA_DOWNLOAD_HP/$MINICONDA_INSTALLER
chmod +x $MINICONDA_INSTALLER
./$MINICONDA_INSTALLER -b -f -p $MINICONDA_PREFIX
```

以下のセルを実行して`conda バージョン`が表示されたらOK

```python
!conda -V
```

Minicondaをアップデート

```python
%%bash
conda init bash
conda update -n base -c defaults conda -y
```

Pythonのバージョンを変更する

```python
!conda install python=$targetPy -y
```
